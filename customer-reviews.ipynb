```python
import pandas as pd
```

# Load dataset

```python
data = pd.read_csv('customer_reviews.csv')
```

# Fill missing values in 'Rating' column with mean

```python

data['Rating'] = data['Rating'].fillna(data['Rating'].mean())
```

# Normalize 'Text Length' column

```python
data['Text Length'] = pd.to_numeric(data['Text Length'], errors='coerce')
data['Text Length'] = data['Text Length'].fillna(data['Text Length'].mean())
data['Text Length'] = (data["Text Length"] - data["Text Length"].mean() / data['Text Length'].std())

print(data.head())
```
