```python
import numpy as np
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.applications.vgg16 import preprocess_input
```

# Load pre-trained VGG16 model + higher level layers

```python
base_model = VGG16(weights='imagenet', include_top=False)
base_model.trainable = False
```

# Add custom top layers for your specific task

```python
num_classes = 10

model = Sequential([
  base_model,
  layers.GlobalAveragePooling2D(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# Create example image data to test the model

```python
num_samples = 200
img_size = 224

X = np.random.rand(num_samples, img_size, img_size, 3).astype('float32')
X = preprocess_input(X)

y = np.random.randint(0, num_classes, size=(num_samples,))

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"y labels: {np.unique(y)}")
```

# Train on the example data

```python
model.fit(X, y, epochs=3, batch_size=16, validation_split=0.2)
```
