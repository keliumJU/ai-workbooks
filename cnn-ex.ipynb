```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow import keras
```

# Building a simple CNN architecture

```python
cnn_model = Sequential([Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
                        MaxPooling2D(pool_size=(2,2)),
                        Conv2D(64, (3,3), activation='relu'),
                        MaxPooling2D(pool_size=(2,2)),
                        Flatten(),
                        Dense(128, activation='relu'),
                        Dense(10, activation='softmax')])

cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

# Load the MNIST dataset

```python
mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```


# Normalize the input data

```python
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
```

# Train CNN on MNIST dataset(reshaping inputs accordingly)

```python
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

cnn_model.fit(x_train_cnn, y_train, epochs=5)
test_loss_cnn, test_acc_cnn = cnn_model.evaluate(x_test_cnn, y_test)
print(f'CNN Test accuracy: {test_acc_cnn}')
```
