```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

# Sample data creation

```python
data = pd.read_csv('transactions.csv')
# Format: 
# 'transaction_id': [1,2,3]
# 'amount': [250.50, 150.75, 75.00]
# 'transaction_date': ['2023-01-15', '2023-01-20', '2023-02-05']

transactions = pd.DataFrame(data)
```

# Convert transaction_data to datetime format

```python

transactions['transaction_date'] = pd.to_datetime(transactions['transaction_date'])
```

# Extract day of the week

```python
transactions['day_of_week'] = transactions['transaction_date'].dt.day_name()
```

# Extract month

```python
transactions['month'] = transactions['transaction_date'].dt.month
```

# Log transformation of amount

```python
transactions['log_amount'] = transactions['amount'].apply(lambda x: np.log(x))
```

# Interaction term between month and amount

```python
transactions['month_amount_interaction'] = transactions['month'] * transactions['amount']
```

# Create a feature for high-value transactions

```python
transactions['high_value'] = transactions['amount'].apply(lambda x: 1 if x > 200 else 0)

print(transactions)
```

# Recursive Feature Elimination(RFE)

```python
X = transactions[['log_amount', 'high_value', 'month_amount_interaction']]
y = transactions['high_value']

model = LogisticRegression()
selector = RFE(model, n_features_to_select=2)
selector = selector.fit(X, y)

print("Selected Features: ", X.columns[selector.support_])
```
