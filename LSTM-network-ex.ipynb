```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

# Assume 'sequences' is your preprocessed text data represented as integers 

```python
vocab_size = 10000 #Example vocabulary size
embedding_dim = 64
```

# Define your LSTM model architecture

```python
lstm_model = Sequential([
  Embedding(input_dim=vocab_size, output_dim=embedding_dim),
  LSTM(units=100),
  Dense(vocab_size, activation='softmax')
])

lstm_model.compile(loss='categorical_crossentropy', optimizer='adam')
```

# Create example sequences to test the model

```python
import numpy as np
import tensorflow as tf

seq_length = 20         # length of each input sequence
num_samples = 1000      # number of training samples

# Generate random integer sequences (in range [0, vocab_size))
input_seqs = np.random.randint(0, vocab_size, size=(num_samples, seq_length))

# Generate random target indices (the "next word" to predict)
target_indices = np.random.randint(0, vocab_size, size=(num_samples,))

# One-hot encode the targets (required for categorical_crossentropy)
targets = tf.keras.utils.to_categorical(target_indices, num_classes=vocab_size)

# Create a tf.data.Dataset that yields (input, target) pairs
sequences = tf.data.Dataset.from_tensor_slices((input_seqs, targets))\
    .batch(32)\
    .prefetch(tf.data.AUTOTUNE)

print(f"Input shape per batch:  (32, {seq_length})")
print(f"Target shape per batch: (32, {vocab_size})")
```

# Train LSTM on your prepared text sequences

```python
lstm_model.fit(sequences, epochs=5)
```
