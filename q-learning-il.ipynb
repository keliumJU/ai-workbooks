```python
import numpy as np
```

# Define parameters

```python
alpha = 0.1 
gamma = 0.9
num_states = 5
num_actions = 2
```

# Initialize Q-table randomly

```python
Q_table = np.random.rand(num_states, num_actions)

def choose_action(state):
  return np.argmax(Q_table[state])

def update_Q(state1, action1, reward1, state2):
  best_future_q = np.max(Q_table[state2])
  Q_table[state1][action1] += alpha * (reward1 + gamma * best_future_q - Q_table[state1][action1])
```

# Example usage(pseudocode)

```python
current_state = 0
action_taken = choose_action(current_state)
reward_received = -1
next_state = 1
update_Q(current_state, action_taken, reward_received, next_state)
```
