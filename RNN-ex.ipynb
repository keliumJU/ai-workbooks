```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

# Prepare your dataset - here we would define our sequences

```python
# Sample text data
texts = [
    'hello world',
    'hello there',
    'world of hello',
    'there and world'
]

# Tokenize texts
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
sequences = pad_sequences(sequences)

vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 50
print(f'Sequences shape: {sequences.shape}')
print(f'Vocab size: {vocab_size}')

# Prepare input (X) and target (y) for next-token prediction
X = sequences[:, :-1]
y = sequences[:, 1:]
print(f'X shape: {X.shape}')
print(f'y shape: {y.shape}')
```

# Define your RNN model architecture

```python
rnn_model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequences.shape[1]-1),
    SimpleRNN(units=100, return_sequences=True),
    Dense(vocab_size, activation='softmax')
])
rnn_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

# Train RNN on your prepared text sequences

```python
rnn_model.fit(X, y, epochs=5)
```
